```python
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans

from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv(r"F:\Mayank Study\Itvedant\DataSet\ML Datasets\Credit Card Customer Data For ClusteringM.csv")
df=df.iloc[:,[2,3]]
df
```

```python
plt.scatter(x=df["Avg_Credit_Limit"],y=df["Total_Credit_Cards"])
```

```python
x=df
x
```

```python
kmeans=KMeans(n_clusters=9)
kmeans.fit(x)
```

```python
pred=kmeans.predict(x)
y=pd.DataFrame(pred)
y
```

```python
plt.scatter(x=df["Avg_Credit_Limit"],y=df["Total_Credit_Cards"],c=y,cmap="rainbow")
```

## Elbow Model

```python
kmeans = KMeans(n_clusters=1)
kmeans.fit(x)
kmeans.predict(x)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
k = np.arange(1,10)
wcss_list = []

for i in k:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
```

```python
wcss_list
```

```python
plt.plot(k,wcss_list)
```
